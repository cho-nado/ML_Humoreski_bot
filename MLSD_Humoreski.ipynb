#Machine Learning Component
Мы строим генератор шуток по модели GPT-3 благодаря доступу к бета-версии OpenAI. GPT-3 использует фреймворк Transformer и имеет обучаемые параметры 175B, 96 слоев, по 96 головок в каждом слое, каждая головка с размерностью 128. Для успешного выполнения генератора шуток GPT-3 требуется всего несколько примеров правильной классификации. Мы собрали датасет, состоящий из 4500 шуток, в этот набор входят анекдоты, забавные диалоги, вопрос-ответ. Модель GPT-3 является мощной моделью генерации текста, поэтому мы взяли именно её. Также важнейшим фактором при выборе нейронной сети, было то, что мы совершенно не знакомы с этим. Знания, которые у нас были - это написание нейронной сети самой обыкновенной, состоящей из 1 слоя. Задачу поставили мы на несколько уровней сложнее, чем опыт, который у нас есть. В связи с этим была выбрана готовая модель.
Мы пытались сделать генератор шуток посредством LSTM системы. LSTM расшифровывается как Long Sкратковременная память.
LSTM — это просто улучшенная версия RNN (Recurrent Neural Nnetworks).
RNN (рекуррентные нейронные сети) работают с последовательными данными.
RNN необходимы, потому что точки данных связаны друг с другом. Следовательно, предыдущие точки данных должны влиять на следующий прогноз, что не может сделать простой MLP.
Вот почему нам нужно передавать информацию из предыдущих точек данных или из памяти в качестве входных данных, что и делает RNN.
Однако при обучении RNN долгосрочные градиенты могут иметь тенденцию к исчезновению (стремление к нулю) или взрыв (стремление к бесконечности), что называется проблемой исчезающего градиента и является причиной того, что стандартные RNN ограниченное.
LSTM — это улучшение RNN, которое частично решает проблему исчезающего градиента.
Ячейка LSTM имеет 3 ворот (ввод, вывод и забвение), которые регулируют поток информации в ячейке.
В чем минус этой системы, что она генерировала несвязанный по смыслу текст, но орфографию и пунктуацию соблюдала. Пришли к выводу, что нужно что-то более мощное. Поэтому GPT-3 идеально подошло.

Примеры явных запросов: "Продолжи шутку, сидели как-то три мужчины еврей, русский, немец в баре за столом и...", "Продолжи шутку, травка зеленеет, солнышко блестит, ласточка с весною...".

Также можно отметить, как нейронная сеть всегда составляет осмысленные предложения с соблюдением пунктуации и орфографии. Юмор прослеживается, даже есть игра слов. Из отрицательного можно подчеркнуть то, что при неявных вопросах о шутках нейронная сеть не всегда понимала, что надо сделать и поэтому продолжала предложение без шуточного аспекта в смысле. Также иногда происходили ошибки с шутками из жизни, бот также продолжал это в форме повествования и просто заканчивал предложение, без шутки. Это решаемо созданием большей базы данных с шутками, на которой бы обучалась нейронная сеть, включив туда больше примеров с каламбурами, смешными ситуациями из жизни, ироничные примеры, саркастические ответы на вопросы. Также было бы особенно важным создать механизм оценки шуток внутри бота после каждой шутки в форме "палец вверх/палец вниз", для создания собственной базы данных, как надо шутить, а как не стоит. Подобная оценка весьма субъективна и разная для каждого человека, поэтому наша нейронная сеть предполагается будет подстраиваться под пользователя, учитывая его предпочтения в юморе.

#Коммерческое обоснование
Целевая аудитория Telegram-бота, генерирующего юмористические истории, может быть достаточно широкой. К ней могут относиться:

* Любители юмора. Это люди, которые любят смеяться и получать удовольствие от юмористических историй.
* Пользователи Telegram. Это люди, которые используют Telegram для общения, получения информации и развлечений.
* Люди, которые хотят отдохнуть и расслабиться. Бот может быть использован для того, чтобы снять стресс и получить хорошее настроение.

5 путей коммерческой реализации Telegram-бота, генерирующего юмористические истории

1. Продажа доступа к боту

Этот способ монетизации является наиболее простым и понятным для пользователей. Пользователи могут покупать подписку на доступ к боту, чтобы получать неограниченное количество каламбурных историй. Стоимость подписки может варьироваться в зависимости от срока подписки, количества историй, которые можно получать в месяц, или других факторов.

Пример:

* Подписка на 1 месяц за 100 рублей
* Подписка на 3 месяца за 250 рублей
* Подписка на 12 месяцев за 600 рублей
---
2. Размещение рекламы в боте

Этот способ монетизации может быть более прибыльным, чем продажа доступа к боту, но он также требует более тщательного планирования. Реклама может размещаться в виде баннеров, нативных объявлений или даже в виде самих историй. Важно, чтобы реклама не была навязчивой и не мешала пользователям получать удовольствие от историй.

Пример:

* Баннеры, которые появляются в самом верху или внизу историй
* Нативные объявления, которые вписываются в контекст историй
* Истории, которые рассказывают о продукте или услуге, которую рекламируют
