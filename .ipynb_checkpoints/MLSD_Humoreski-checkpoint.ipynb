# Machine Learning Component
We are building a joke generator based on the GPT-3 model thanks to access to the OpenAI beta version. GPT-3 uses the Transformer framework and has 175 billion trainable parameters, 96 layers, 96 heads in each layer, and each head with a dimension of 128. To successfully run the GPT-3 joke generator, only a few examples of correct classification are needed. We collected a dataset of 4,500 jokes, including anecdotes, funny dialogues, and question-answer formats. GPT-3 was chosen because it is a powerful text generation model. Another crucial factor in choosing this neural network was our lack of familiarity with such models. Our previous experience was limited to writing a simple neural network consisting of just one layer. We set ourselves a task far more complex than our experience allowed, which is why we opted for a pre-built model.

We initially attempted to create a joke generator using an LSTM system. LSTM stands for Long Short-Term Memory.

LSTM is simply an improved version of RNN (Recurrent Neural Networks). RNNs work with sequential data. RNNs are necessary because data points are interrelated, meaning that previous data points should influence the next prediction, something that a simple MLP cannot achieve. This is why information from previous data points, or memory, must be passed along as input, which is what RNNs do.

However, when training RNNs, long-term gradients tend to vanish (approaching zero) or explode (approaching infinity), which is known as the vanishing gradient problem, a limitation of standard RNNs. LSTM improves upon RNNs by partially solving the vanishing gradient issue.

An LSTM cell has three gates (input, output, and forget) that regulate the flow of information within the cell.

The downside of this system was that it generated text that lacked coherence, although it adhered to proper spelling and punctuation. We concluded that something more powerful was needed, which made GPT-3 the ideal choice.


# System Evaluation
To assess the performance of our neural network, we integrated it and connected it to the bot we created in the social network Telegram, simulating multiple user interactions with the bot over several sessions. We set up a dedicated server for the bot so it could operate not only on a local device but also be available to anyone at any time. We tested the system using various types of queries, differing both semantically and in meaning—personal and general. We also included both implicit and explicit types of questions—those directly related to jokes. Examples of implicit queries: "Once, three men, a Jew, a Russian, and a German were sitting at a bar, and...", "The grass is green, the sun is shining, the swallow is returning with the spring...". Examples of explicit queries: "Continue the joke, once, three men, a Jew, a Russian, and a German were sitting at a bar, and...", "Continue the joke, the grass is green, the sun is shining, the swallow is returning with the spring...".

We also evaluated the latency and throughput of our system in generating responses to queries of varying complexity, as well as when generating responses for different users simultaneously using the bot with the neural network.

Additionally, we asked our friends and colleagues to interact with HumoreskiBot and provide feedback based on the bot’s responses to their queries—most feedback was positive, especially regarding performance. It was also noted how the neural network always generates meaningful sentences with correct grammar and punctuation. Humor is present, and there is even some wordplay.

On the negative side, it was pointed out that the neural network sometimes didn’t understand implicit joke-related queries, continuing the sentence without adding any humorous aspect. There were also issues with jokes based on real-life situations, where the bot continued in narrative form, simply completing the sentence without delivering a punchline. This could be solved by expanding the joke database on which the neural network is trained, including more examples of puns, funny life situations, ironic examples, and sarcastic responses to questions. It would also be beneficial to implement a joke evaluation mechanism within the bot after each joke, in the form of a "thumbs up/thumbs down" system, to help build a database of what works as a joke and what doesn’t. Such evaluation is highly subjective and varies from person to person, so our neural network is expected to adapt to each user’s humor preferences.

In future development, once we have a larger user base, we will be able to conduct more in-depth data analysis. For example, we can track user experience metrics (such as the average number of queries per session and the total number of simultaneously active sessions) and bot quality metrics (such as the most frequent responses, the percentage of inappropriate answers where the chatbot failed to respond due to a lack of context or failed to include any humor).

```python
import random
import telebot
import glob
from llama_index import StorageContext, load_index_from_storage
import requests
import os

os.environ["OPENAI_API_KEY"] = "YOUR_OPENAI_CODE"
storage_context = StorageContext.from_defaults(persist_dir="./entire_docs")
index = load_index_from_storage(storage_context)
query_engine = index.as_query_engine()

bot = telebot.TeleBot("YOUR_TELEGRAM_TOKEN")

@bot.message_handler(commands=['start'])
def send_random_image(message):
    bot.reply_to(message, "Hello, I am the bot!")
    image_folder_path = "./memes_random"  # specify the path to the folder with pictures

    image_files = glob.glob(image_folder_path + '/*.jpg') + glob.glob(image_folder_path + '/*.png')

    if image_files:
        random_image = random.choice(image_files)
        with open(random_image, "rb") as image_file:
            bot.send_photo(chat_id=message.chat.id, photo=image_file)
    else:
        bot.reply_to(message, "There is no pictures in the folder. ")

@bot.message_handler(func=lambda message: True)
def echo_all(message):
    text_input = message.text
    text_input = "Continue the joke: " + text_input
    response = query_engine.query(text_input)
    bot.reply_to(message, "Bot: {}".format(response))

bot.polling()
```
